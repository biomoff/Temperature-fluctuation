# Library Imports

```python
# For data manipulation
import pandas as pd
import numpy as np
```

---

# Load Trait Data

```python
# Sample metadata
sample_data = pd.read_csv("../Datasets/sample-data.csv")

# Temperature optima
optima = pd.read_csv("../Datasets/Optimum_temperature/Optima.csv")
asv_list = optima['ASV_ID']

# Niche breadth
niche_breadth = pd.read_csv("../Datasets/Niche_breadth/StageOneActive_NicheBreadth.csv", index_col= 0)
niche_breadth = niche_breadth.reset_index(names = ['ASV_ID'])
niche_breadth = niche_breadth[['ASV_ID','Bn']] # remove unwanted columns
niche_breadth = niche_breadth[niche_breadth['ASV_ID'].isin(asv_list)] # subset to ASVs of interest

# Fitness
highT_fitness = pd.read_csv("../Datasets/ALDEx2/highT_ALDEx2.csv", index_col= 0)
highT_fitness = highT_fitness.reset_index(names = ['ASV_ID'])
highT_fitness = highT_fitness[['ASV_ID', 'effect']] # remove unwanted columns
highT_fitness = highT_fitness.rename(columns= {'effect': 'highT_fitness'}) # rename columns for merging
highT_fitness = highT_fitness[highT_fitness['ASV_ID'].isin(asv_list)] # subset to ASVs of interest

lowT_fitness = pd.read_csv("../Datasets/ALDEx2/lowT_ALDEx2.csv", index_col= 0)
lowT_fitness = lowT_fitness.reset_index(names = ['ASV_ID'])
lowT_fitness = lowT_fitness[['ASV_ID', 'effect']] # remove unwanted columns
lowT_fitness = lowT_fitness.rename(columns= {'effect': 'lowT_fitness'}) # rename columns for merging
lowT_fitness = lowT_fitness[lowT_fitness['ASV_ID'].isin(asv_list)] # subset to ASVs of interest

TwoWeeks_fitness = pd.read_csv("../Datasets/ALDEx2/TwoWeeks_ALDEx2.csv", index_col= 0)
TwoWeeks_fitness = TwoWeeks_fitness.reset_index(names = ['ASV_ID'])
TwoWeeks_fitness = TwoWeeks_fitness[['ASV_ID', 'effect']] # remove unwanted columns
TwoWeeks_fitness = TwoWeeks_fitness.rename(columns= {'effect': 'TwoWeeks_fitness'}) # rename columns for merging
TwoWeeks_fitness = TwoWeeks_fitness[TwoWeeks_fitness['ASV_ID'].isin(asv_list)] # subset to ASVs of interest

OneWeek_fitness = pd.read_csv("../Datasets/ALDEx2/OneWeek_ALDEx2.csv", index_col= 0)
OneWeek_fitness = OneWeek_fitness.reset_index(names = ['ASV_ID'])
OneWeek_fitness = OneWeek_fitness[['ASV_ID', 'effect']] # remove unwanted columns
OneWeek_fitness = OneWeek_fitness.rename(columns= {'effect': 'OneWeek_fitness'}) # rename columns for merging
OneWeek_fitness = OneWeek_fitness[OneWeek_fitness['ASV_ID'].isin(asv_list)] # subset to ASVs of interest

TwoDays_fitness = pd.read_csv("../Datasets/ALDEx2/TwoDays_ALDEx2.csv", index_col= 0)
TwoDays_fitness = TwoDays_fitness.reset_index(names = ['ASV_ID'])
TwoDays_fitness = TwoDays_fitness[['ASV_ID', 'effect']] # remove unwanted columns
TwoDays_fitness = TwoDays_fitness.rename(columns= {'effect': 'TwoDays_fitness'}) # rename columns for merging
TwoDays_fitness = TwoDays_fitness[TwoDays_fitness['ASV_ID'].isin(asv_list)] # subset to ASVs of interest
```

---

# Calculate Niche Position

```python
optima['NP_highT'] = abs(30 - optima['Optimum'])
optima['NP_lowT'] = abs(15 - optima['Optimum'])
optima['NP_fluctuation'] = abs(22.5 - optima['Optimum'])
optima['NP_TwoWeeks'] = abs(20 - optima['Optimum'])
```

---

# Merge All Trait Data

```python
# Merge in temperature optima and niche breadth
trait_data = pd.merge(left= optima, right= niche_breadth, how= 'left', on= 'ASV_ID')

# Merge in fold change in 
## High temperature stable
trait_data = pd.merge(left= trait_data, right= highT_fitness, how= 'left', on= 'ASV_ID')

## Low temperature stable
trait_data = pd.merge(left= trait_data, right= lowT_fitness, how= 'left', on= 'ASV_ID')

## Two week fluctuation
trait_data = pd.merge(left= trait_data, right= TwoWeeks_fitness, how= 'left', on= 'ASV_ID')

## One week fluctuation
trait_data = pd.merge(left= trait_data, right= OneWeek_fitness, how= 'left', on= 'ASV_ID')

## Two day fluctuation
trait_data = pd.merge(left= trait_data, right= TwoDays_fitness, how= 'left', on= 'ASV_ID')
```

# Export To File

```python
# Drop any ASVs that do not have data for all traits
trait_data = trait_data.dropna(axis= 0)

trait_data.to_csv("../Datasets/ASV_trait_data.csv", index=False)
```
