# Convert Relative To Absolute Abundance Using qPCR Data

```python
def absolute_abundance(data, qPCR):
    # Merge relative abundance table with qPCR data
    data = pd.merge(left= data, right= qPCR[['Sample', 'qPCR']], how= 'left', on= 'Sample' )
    
    # Calculate absolute abundance
    data['Absolute_abundance'] = data['Count'] * data['qPCR']
    data.drop('qPCR', axis= 1, inplace= True)
    
    # Rename 'Count' columun to 'Relative_abundance'
    data.rename(columns= {'Count': 'Relative_abundance'}, inplace= True)
    return data
```

```python
# Calculate absolute abundance for 
s1_active_absolute = absolute_abundance(s1_active_relative, qPCR) # Stage 1 Active Data
s2_active_absolute = absolute_abundance(s2_active_relative, qPCR) # Stage 2 Active Data
s1_resident_w8_absolute = absolute_abundance(s1_resident_w8_relative, qPCR) # Stage 1 Resident Data at Week 8
s2_resident_w8_absolute = absolute_abundance(s2_resident_w8_relative, qPCR) # Stage 2 Resident Data at Week 8
s2_resident_absolute = absolute_abundance(s2_resident_relative, qPCR) # Stage 2 Resident Data for all Time Points

# Export to Files
s1_active_absolute.to_csv("../Datasets/Absolute_abundance/StageOneActive_AA_long.csv", index= False)
s2_active_absolute.to_csv("../Datasets/Absolute_abundance/StageTwoActive_AA_long.csv", index= False)
s1_resident_w8_absolute.to_csv("../Datasets/Absolute_abundance/StageOneResidentWeekEight_AA_long.csv", index= False)
s2_resident_w8_absolute.to_csv("../Datasets/Absolute_abundance/StageTwoResidentWeekEight_AA_long.csv", index= False)
s2_resident_absolute.to_csv("../Datasets/Absolute_abundance/StageTwoResident_AA_long.csv", index= False)
```
